```python
import os
import shutil

# Original dataset path
dataset_path = "/content/drive/MyDrive/LungCancerProject/data/The IQ-OTHNCCD lung cancer dataset/The IQ-OTHNCCD lung cancer dataset"

# New binary dataset path
binary_path = "/content/drive/MyDrive/LungCancerProject/data_binary"

# Create folders
os.makedirs(os.path.join(binary_path, "Cancer"), exist_ok=True)
os.makedirs(os.path.join(binary_path, "Normal"), exist_ok=True)

# Merge Benign + Malignant into Cancer
for folder in ["Bengin cases", "Malignant cases"]:
    folder_path = os.path.join(dataset_path, folder)
    for img in os.listdir(folder_path):
        shutil.copy(
            os.path.join(folder_path, img),
            os.path.join(binary_path, "Cancer", img)
        )

# Copy Normal cases
normal_path = os.path.join(dataset_path, "Normal cases")
for img in os.listdir(normal_path):
    shutil.copy(
        os.path.join(normal_path, img),
        os.path.join(binary_path, "Normal", img)
    )

print("Binary dataset created successfully!")
```

```python
for cls in ["Cancer", "Normal"]:
    path = os.path.join(binary_path, cls)
    print(cls, ":", len(os.listdir(path)))
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

binary_path = "/content/drive/MyDrive/LungCancerProject/data_binary"
split_path = "/content/drive/MyDrive/LungCancerProject/data_binary_split"

classes = ["Cancer", "Normal"]
splits = ["train", "val", "test"]

# Create split folders
for split in splits:
    for cls in classes:
        os.makedirs(os.path.join(split_path, split, cls), exist_ok=True)

# Split data
for cls in classes:
    cls_path = os.path.join(binary_path, cls)
    images = os.listdir(cls_path)

    train_imgs, temp_imgs = train_test_split(images, test_size=0.3, random_state=42)
    val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, random_state=42)

    for img in train_imgs:
        shutil.copy(os.path.join(cls_path, img),
                    os.path.join(split_path, "train", cls, img))

    for img in val_imgs:
        shutil.copy(os.path.join(cls_path, img),
                    os.path.join(split_path, "val", cls, img))

    for img in test_imgs:
        shutil.copy(os.path.join(cls_path, img),
                    os.path.join(split_path, "test", cls, img))

print("Train/Val/Test split complete!")
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

split_path = "/content/drive/MyDrive/LungCancerProject/data_binary_split"

# Image transformations
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

val_test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# Datasets
train_dataset = datasets.ImageFolder(split_path + "/train", transform=train_transform)
val_dataset = datasets.ImageFolder(split_path + "/val", transform=val_test_transform)
test_dataset = datasets.ImageFolder(split_path + "/test", transform=val_test_transform)

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)

print("Classes:", train_dataset.classes)
```

```python
import torch
import torch.nn as nn
from torchvision import models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load pretrained ResNet50
model = models.resnet50(pretrained=True)

# Freeze early layers (optional but good for medical imaging)
for param in model.parameters():
    param.requires_grad = True  # keep trainable for now

# Replace final layer
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 2)

model = model.to(device)

print("Model loaded successfully!")
```

```python
# Class counts
cancer_count = 681
normal_count = 416

total = cancer_count + normal_count

# Inverse frequency
weight_cancer = total / cancer_count
weight_normal = total / normal_count

class_weights = torch.tensor([weight_cancer, weight_normal]).to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)

print("Loss & Optimizer ready!")
```

```python
print(class_weights)
```

```python
import torch
from tqdm import tqdm

num_epochs = 10

best_val_acc = 0.0
model_save_path = "/content/drive/MyDrive/LungCancerProject/models/best_model.pth"

for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")

    # -------------------
    # TRAINING
    # -------------------
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in tqdm(train_loader):
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    train_acc = 100 * correct / total
    train_loss = running_loss / len(train_loader)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train Accuracy: {train_acc:.2f}%")

    # -------------------
    # VALIDATION
    # -------------------
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_acc = 100 * correct / total
    print(f"Validation Accuracy: {val_acc:.2f}%")

    # Save best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), model_save_path)
        print("Best model saved!")

print("\nTraining Complete!")
```

```python
import torch
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Load best model
model.load_state_dict(torch.load(model_save_path))
model.eval()

all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Accuracy
test_acc = 100 * np.mean(np.array(all_preds) == np.array(all_labels))
print("Test Accuracy: {:.2f}%".format(test_acc))

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
print("\nConfusion Matrix:")
print(cm)

# Classification Report
print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=["Cancer", "Normal"]))
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Load best model
model.load_state_dict(torch.load(model_save_path))
model.eval()

all_probs = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        probs = torch.softmax(outputs, dim=1)

        # Probability of class "Cancer" (index 0)
        all_probs.extend(probs[:, 0].cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

all_probs = np.array(all_probs)
all_labels = np.array(all_labels)

# Compute ROC
fpr, tpr, thresholds = roc_curve(all_labels, all_probs, pos_label=0)
roc_auc = auc(fpr, tpr)

print("AUC Score:", roc_auc)

# Plot ROC
plt.figure()
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```
